Abaixo temos a configuração do ambiente SQL:

```python
import sqlite3
import pandas as pd
```

```python
df_transacoes = pd.read_csv("TB_TRANSACOES_PROJETO_ECOMM.csv", delimiter=';')
df_clientes = pd.read_csv("TB_CLIENTES_PROJETO_ECOMM.csv", delimiter=';')
```

```python
conn = sqlite3.connect('projeto.db')
# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
unic_trans = df_transacoes['id_client'].nunique()
unic_client = df_clientes['Id_client'].nunique()

print(f"Valores unicos em transações:{unic_trans} \n Valores únicos em clientes: {unic_client}")
```

```python
query = """
SELECT
A.*,
B.*
FROM tb_transacoes as A
INNER JOIN tb_clientes as B
ON A.Id_client = B.id_client
"""
result_df = run_query(query)
print(result_df)
```

Justifique a escolha do JOIN:

O INNER JOIN foi escolhido pois a tabela de transações apresenta dados de multiplas transações realizadas por alguns clientes além de outras transações de clientes que não possuem cadastro na tabela de clientes. Dessa forma, o inner join pega apenas os id de clientes que realizam transações e possuem cadastro além de manter os dados de multiplas transações do mesmo cliente que são importantes para análises mais complexas.

Exportando o arquivo como CSV:

```python
result_df.to_csv('dados_ecommerce_final.csv', index=False)
```
